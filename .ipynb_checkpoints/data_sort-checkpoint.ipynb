```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
df = pd.read_excel('data/hist_data_nm_cat.xlsx')
df = df.rename(columns={'day': 'date'})
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.drop(['month'], axis=1, inplace=True)
df
```

```python
# check period
m_names = list(set(df['merchant_name'].tolist()))
df['id'] = 0
nid = 0
df2 = df.copy()[0:0]

for n in m_names:
    data_by_name = df[df['merchant_name'] == n]
    dbn_i = data_by_name.index
    for j in range(len(dbn_i)-1):
        data_by_name.loc[dbn_i[j], 'id'] = nid
        if (dbn_i[j+1] - dbn_i[j]).days > 1:
            nid+=1
    data_by_name.loc[dbn_i[-1], 'id'] = nid
    df2 = pd.concat([df2, data_by_name])
    nid+=1 
df2
```

```python
df2.to_csv('data/train_data.csv');
```

```python
ids_list = list(set(df2['id']))
```

```python
datal = {}

for i in ids_list:
    count = len(df2[df2['id']==i])
    budget = int(df2.loc[df2['id'] == i, 'cashback'].sum())
    if (count > 100) and (budget > 5000):
        datal[i] = [count, budget, df2.loc[df2['id'] == i, 'merchant_name'].iloc[0]]
        

datal = dict(sorted(datal.items(), key=lambda item: item[1], reverse=True))
datal
```
