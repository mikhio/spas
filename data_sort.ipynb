```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/hist_data_nm_cat.xlsx')
df = df.rename(columns={'day': 'date'})
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.drop(['month'], axis=1, inplace=True)
df
```

```python
# check period
m_names = list(set(df['merchant_name'].tolist()))
df['id'] = 0
nid = 0
df2 = df.copy()[0:0]

for n in m_names:
    data_by_name = df[df['merchant_name'] == n]
    dbn_i = data_by_name.index
    for j in range(len(dbn_i)-1):
        data_by_name.loc[dbn_i[j], 'id'] = nid
        if (dbn_i[j+1] - dbn_i[j]).days > 1:
            nid+=1
    data_by_name.loc[dbn_i[-1], 'id'] = nid
    df2 = pd.concat([df2, data_by_name])
    nid+=1 
df2
```
